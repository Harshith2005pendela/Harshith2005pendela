# **Parameters**

```python
#parameters
#discount factor
gamma = 0.99
#number of episodes
n_episodes = 1000
#number of time steps
n_timesteps = 1000
```

# **Defining the Neural network for the model**



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, BatchNormalization
from tensorflow.keras.optimizers import RMSprop
import tensorflow.keras.backend as K

%matplotlib inline
```

```python

model = Sequential()
model.add(Conv2D(4, kernel_size=(3,3), padding='same', activation='relu', input_shape = (80,80,1)))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Conv2D(8, kernel_size=(3,3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Conv2D(12, kernel_size=(3,3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Conv2D(16, kernel_size=(3,3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(len(action_space), activation='softmax'))

model.compile(optimizer='rmsprop', loss='sparse_categorical_crossentropy')

model.summary()
```

# **Preprocessing**

```python
# preprocess the image into a 80*80*1 image
def preprocess(I):
    I = I[35:195] # crop
    I = I[::2,::2,0] # downsample by factor of 2
    I[I == 144] = 0 # erase background (background type 1)
    I[I == 109] = 0 # erase background (background type 2)
    I[I != 0] = 1 # everything else (paddles, ball) just set to 1
    return I.astype(float)[:,:,None]
```

```python
total_reward = np.zeros(n_episodes)
losses = np.zeros(n_episodes)
time_taken = np.zeros(n_episodes)
epi_reward = 0
converted_shape = (80, 80, 1)

prev_frame = None
# remembering previous experiences(It's storing 1000 frames /state transitions)[1000 time steps]
buffer = 1000
xs = np.zeros((buffer,)+converted_shape)
ys = np.zeros((buffer,1))
rs = np.zeros((buffer))
```

```python
def discount(r):
    # It is nitialized to the same shape as r (the input rewards array), but with all elements set to zero.
    discounted_r = np.zeros_like(r)
    # This is the value which we want to change and assign it to different elements in the list
    running_add = 0
    # As the discounted reward lessers the reward in future we have to start from the future rewards and then come to the present rewards
    for t in reversed(range(len(discounted_r))):
      # In pong reward of win is +1 and loss is -1 so whenever it is non-zero it marks the end of a round in the episode and after the end of episode the new reward should start again so assigning running_add to 0.
        if r[t] != 0: running_add = 0
        running_add =  r[t] + running_add * gamma
        discounted_r[t] = running_add
    return discounted_r
def standardize(r):
  dr = discount(r)
  dr = (dr - dr.mean(r)) / (dr.std(r))
  return dr
```

```python
k = 0
for episodes in range(n_episodes):
  state = env.reset()[0]
  # Get the current state of environment
  x = preprocess(state)
  #focussing on difference between frames instead of frames as that is only important for our agent to learn
  xs[k] = x - prev_frame if prev_frame is not None else np.zeros(converted_shape)
  # store the current state as prev_state for the next iteration
  prev_frame = x

  # probability distribution by the actions which we took
  p = model.predict(xs[k][None,:,:,:])
  # picking a random action with p[0] being the list of the probabilities associated with every action
  a = np.random.choice(len(action_space), p=p[0])
  # choosing action
  action = action_space[a]
  # storing action
  ys[k] = a

  # taking that action , we store the reward
  next_state, reward, done,info = env.step(action)
  # episode reward is the total reward our agent get's in the episode
  epi_reward += reward
  # reward per time step
  rs[k] = reward

  k += 1
  # If the episode get's completed or 1000 time steps are reached
  if done or k==buffer:
  # array which stores the rewards of all episodes
    total_reward[episodes] = epi_reward
  # after one episode reward is set to get the reward of next episode I set the epi_reward to be 0
    epi_reward = 0
  # get the  state, action (y), and rewards (and preprocess)
  # ep_x collects all the states which were observed in that episode till the time frame k(k not included)
    ep_x = xs[:k]
  # ep_y collects all the actions which were observed in that episode till the time frame k(k not included)
    ep_y = ys[:k]
  # ep_r collects all the intermediate rewards which were observed in that episode till the time frame k(k not included)
    ep_r = rs[:k]
  # discount rewards , In RL we give preference to discounted rewards which states that reward that I get now is slightly/ highly wanted by me for many reasons
    ep_r = discount(ep_r)
  # fitting ML Model with the data which we have states info , actions taken during that state and a Measure of how good it is to be in that state by Q value(Monte Carlo)
  # verbose is taken as zero because we don't want to observe so much of output and we want a brief and crisp output which is important
    model.fit(ep_x, ep_y, sample_weight=ep_r, batch_size=512, epochs=1, verbose=0)
    # gets the number of time steps spent in the current episode
    time_taken[episodes] = k
  # reset the buffer
    k = 0
  # reset the previous frame to none like a new start
    prev_frame = None
  # reset the environment
    state = env.reset()[0]
  # calculating losses and storing it in losses array
    losses[episodes] = model.evaluate(ep_x,
                                      ep_y,
                                      sample_weight=ep_r,
                                      batch_size=len(ep_x),
                                      verbose=0)
    print(episodes)
      # Print out metrics like rewards, how long each episode lasted etc.
    if episodes%(n_episodes//20) == 0:
      # getting the average reward over 200 episodes
        avg_reward = np.mean(total_reward[max(0,episodes-200):episodes])
      # getting the average loss for 200 episodes
        avg_loss = np.mean(losses[max(0,episodes-200):episodes])
      # getting the average time for 200 episodes
        avg_time = np.mean(time_taken[max(0,episodes-200):episodes])
        print('Episode: {0:d}, Average Loss: {1:.4f}, Average Reward: {2:.4f}, Average steps: {3:.4f}'
              .format(episodes, avg_loss, avg_reward, avg_time))
```
