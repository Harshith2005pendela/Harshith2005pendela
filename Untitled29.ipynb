```python
import numpy as np
import matplotlib.pyplot as plt

# Problem parameters (4-armed bandit)
means_a = [2.0, 1.8, 1.5, 1.0]
optimal_mean_a = max(means_a)  # Optimal mean = 2.0
k = len(means_a)               # Number of arms (k=4)
T = 2000                       # Time horizon
m_values = [10, 20, 50, 100, 200]  # ETC exploration values
```

```python
# ETC Algorithm
def run_etc(m, T, means):
    k = len(means)
    explore_steps = m * k
    if explore_steps > T:
        raise ValueError("m is too large for T")
    emp_means = np.zeros(k)
    n = np.zeros(k)
    regret = np.zeros(T)
    t = 0

    # Explore phase
    for _ in range(m):
        for arm in range(k):
            reward = np.random.normal(means[arm], 1)
            emp_means[arm] = (emp_means[arm] * n[arm] + reward) / (n[arm] + 1)
            n[arm] += 1
            regret[t] = optimal_mean_a - means[arm]
            t += 1

    # Commit phase
    best_arm = np.argmax(emp_means)
    while t < T:
        regret[t] = optimal_mean_a - means[best_arm]
        t += 1
    return np.cumsum(regret)
```

```python
def run_ucb(T, means):
    k = len(means)
    n = np.ones(k)  # Start with 1 pull per arm
    sum_rewards = np.zeros(k)
    cumulative_regret = np.zeros(T)
    total_regret = 0

    # Initial pulls (1 per arm)
    for arm in range(k):
        reward = np.random.normal(means[arm], 1)
        sum_rewards[arm] = reward
        inst_regret = optimal_mean_a - means[arm]
        total_regret += inst_regret
        cumulative_regret[arm] = total_regret

    # UCB with log(T) and alpha=2 (textbook)
    for t in range(k, T):
        ucb = sum_rewards / n + np.sqrt(4 * np.log(T)) / n
        arm = np.argmax(ucb)
        reward = np.random.normal(means[arm], 1)
        sum_rewards[arm] += reward
        n[arm] += 1
        inst_regret = optimal_mean_a - means[arm]
        total_regret += inst_regret
        cumulative_regret[t] = total_regret

    return cumulative_regret
```

```python
def run_anytime_ucb(T, means):
    k = len(means)
    n = np.ones(k)  # Start with 1 pull per arm
    sum_rewards = np.zeros(k)
    cumulative_regret = np.zeros(T)
    total_regret = 0

    # Initial pulls (1 per arm)
    for arm in range(k):
        reward = np.random.normal(means[arm], 1)
        sum_rewards[arm] = reward
        inst_regret = optimal_mean_a - means[arm]
        total_regret += inst_regret
        cumulative_regret[arm] = total_regret

    # UCB with log(t) and alpha=1 (anytime)
    for t in range(k, T):
        ucb = sum_rewards / n + np.sqrt(np.log(t) / n)  # alpha=1, log(t)
        arm = np.argmax(ucb)
        reward = np.random.normal(means[arm], 1)
        sum_rewards[arm] += reward
        n[arm] += 1
        inst_regret = optimal_mean_a - means[arm]
        total_regret += inst_regret
        cumulative_regret[t] = total_regret

    return cumulative_regret
```

```python
# Initialize regret storage
num_runs = 100              # Number of simulation runs

etc_regrets = {m: np.zeros(T) for m in m_values}
ucb_regrets = np.zeros(T)
anytime_ucb_regrets = np.zeros(T)

# Run simulations
for run in range(num_runs):
    print(f"Run {run+1}/{num_runs}")

    # ETC (skip if exploration exceeds T)
    for m in m_values:
        if k * m > T:
            continue
        etc_regret = run_etc(m, T, means_a)
        etc_regrets[m] += etc_regret

    # Standard UCB
    ucb_regret = run_ucb(T, means_a)
    ucb_regrets += ucb_regret

    # Anytime UCB
    anytime_regret = run_anytime_ucb(T, means_a)
    anytime_ucb_regrets += anytime_regret

# Average results
for m in m_values:
    if k * m <= T:
        etc_regrets[m] /= num_runs
ucb_regrets /= num_runs
anytime_ucb_regrets /= num_runs

# Plotting
plt.figure(figsize=(10, 6))
for m in m_values:
    if k * m <= T:
        plt.plot(etc_regrets[m], label=f'ETC m={m}')
plt.plot(ucb_regrets, label='Standard UCB (α=2, log(T))')
plt.plot(anytime_ucb_regrets, label='Anytime UCB (α=1, log(t))')
plt.xlabel('Time Step')
plt.ylabel('Average Cumulative Regret')
plt.title('Regret Comparison: ETC vs Standard UCB vs Anytime UCB')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Part (b): Varying µ2 in [1.55, 1.95]
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 2000
num_runs = 1000
mu2_values = np.linspace(1.55, 1.95, 15)  # 15 points between 1.55 and 1.95
avg_final_regrets = []

# Fixed means (only µ2 varies)
mu1 = 2.0
mu3 = 1.5
mu4 = 1.0

# Run simulations for each µ2
for mu2 in mu2_values:
    current_means = [mu1, mu2, mu3, mu4]
    optimal_mean = max(current_means)
    total_regret = 0.0

    for run in range(num_runs):
        # Run Anytime UCB
        cumulative_regret = run_anytime_ucb(T, current_means)
        total_regret += cumulative_regret[-1]  # Regret at final timestep T

    avg_final_regret = total_regret / num_runs
    avg_final_regrets.append(avg_final_regret)
    print(f"µ2 = {mu2:.2f}, Avg Regret = {avg_final_regret:.2f}")

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(mu2_values, avg_final_regrets, 'b-', marker='o', linewidth=2)
plt.xlabel('µ2 Value', fontsize=12)
plt.ylabel('Average Final Cumulative Regret (T=2000)', fontsize=12)
plt.title('Anytime UCB Performance: Varying µ2 in [1.55, 1.95]', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(np.arange(1.55, 2.0, 0.05))  # Show ticks every 0.05
plt.show()
```
